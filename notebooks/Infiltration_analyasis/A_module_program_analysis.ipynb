```python
sample = 'B926-1'
proportions = []

for i in range(4):
    mask = cnt_matrix_trans[fdata.obs.marker == sample, i] > np.quantile(cnt_matrix_trans[fdata.obs.marker == sample, i], 0.75)
    df = pd.DataFrame(np.array(gdata[gdata.obs.marker == sample, :][mask, :].X.sum(axis=0).flatten()), index=gdata.var_names)
    df.columns = ['guide']
    df['mask'] = f'Mask {i}'
    proportions.append(df)

proportions_df = pd.concat(proportions)
proportions_df.reset_index(inplace=True)
proportions_df.rename(columns={'index': 'guide_name'}, inplace=True)
proportions_df['proportion'] = proportions_df.groupby('guide_name')['guide'].transform(lambda x: x / x.sum())

proportions_pivot = proportions_df.pivot(index='guide_name', columns='mask', values='proportion')
proportions_pivot = proportions_pivot[['Mask 0', 'Mask 1', 'Mask 2', 'Mask 3']]

# 按照Mask2的Proportion由小到大排列
proportions_pivot = proportions_pivot.sort_values(by='Mask 3')

# 更改Legend的名字
cmap = sns.color_palette("Reds", as_cmap=True)
cmap = cmap(np.linspace(0.25, 0.75, 4))  # 只取>0.5的部分
# fig = proportions_pivot.plot(kind='bar', stacked=True, figsize=(10, 3), color=cmap, width=1)  # 设置bar的宽度

# 拟合每一种Program并绘制不同颜色的线
for i in range(4):
    plt.plot(proportions_pivot.index, proportions_pivot.iloc[:, i], color=cmap[i], label=f'Mask {i} Fit', linewidth=2)

plt.xticks(rotation=90)
plt.legend(['Infiltration', 'Fibro', 'Tumor', 'Necrosis'] + [f'Mask {i} Fit' for i in range(4)], bbox_to_anchor=(1.05, 1), loc='upper left',
           facecolor='white', fontsize='small', frameon=False)
plt.title('Enrichment by NMF Program', color='black')  # 设置标题字体为白色
sns.despine()
plt.show()
```

```python
# 选择需要的guides
guides_of_interest = ['sgCd44', 'sgGata3', 'sgnon-targeting', 'sgCxcl16']

df_list = []
for sample in ['1-1', '1-2', '2-1', '2-2', '3-1', '3-2']:
    df = gdata_bin100[gdata_bin100.obs.marker == sample, guides_of_interest].X
    df = pd.DataFrame(df, index=gdata_bin100[gdata_bin100.obs.marker == sample, guides_of_interest].obs_names, columns=guides_of_interest)
    df = pd.concat([df, gdata_bin100[gdata_bin100.obs.marker == sample, :].obs.cluster], axis=1)
    df.cluster = df.cluster.astype(str)
    df = df.melt(id_vars=['cluster'], value_vars=guides_of_interest, var_name='guide', value_name='value')
    df.value = df.value.astype(float)
    # 首先计算每个guide的总和
    guide_sums = df.groupby('cluster')['value'].sum()
    # 然后计算每个cluster和guide组合的和,并除以对应guide的总和
    df = df.groupby(['cluster', 'guide'])['value'].sum().reset_index()
    df['value'] = df.apply(lambda x: x['value'] / guide_sums[x['cluster']], axis=1)
    df['sample'] = sample
    df_list.append(df)
df_all = pd.concat(df_list)
df_all
```

```python
plot_df_all = []
for sample in ['1-1', '1-2', '2-1', '2-2', '3-1', '3-2']:
    # 获取指定sample的数据
    df_all_sample = df_all.loc[df_all['sample'] == sample, :].copy()

    # 计算每个cluster的NMF mean score
    nmf_scores = fdata_bin100.obs.groupby('cluster_cellcharter_givenk')[['Mask 0', 'Mask 1', 'Mask 2', 'Mask 3', 'Mask 4', 'Mask 5']].mean()

    # 对每一列进行归一化处理
    nmf_scores_normalized = nmf_scores.apply(lambda x: (x - x.mean()) / np.sqrt(x.var()))

    # 计算每个guide在每个cluster的proportion和NMF score的乘积
    results = []
    for cluster in df_all_sample.cluster.unique():
        cluster_data = df_all_sample[df_all_sample.cluster == cluster]
        
        # 获取该cluster的NMF scores
        cluster_nmf_scores = nmf_scores_normalized.loc[int(cluster)]
        
        for guide in cluster_data.guide.unique():
            guide_value = cluster_data[cluster_data.guide == guide].value.values[0]
            
            # 计算每个program的相关性得分
            for i in range(6):
                nmf_score = cluster_nmf_scores[f'Mask {i}']
                correlation = guide_value * nmf_score      
                results.append({
                    'cluster': cluster,
                    'guide': guide,
                    'program': i,
                    'correlation_score': correlation
                })

    # 转换为DataFrame
    correlation_df = pd.DataFrame(results)
    plot_df = correlation_df.groupby(['program', 'guide'])['correlation_score'].mean().unstack()
    plot_df = plot_df.melt(ignore_index=False).reset_index(drop=False)
    plot_df['sample'] = sample

    plot_df_all.append(plot_df)

plot_df_all = pd.concat(plot_df_all)
plot_df_all
```

```python
# 计算每个guide在network中与其他program的互作的平均值
network_df = plot_df_all.loc[plot_df_all['sample'] == '2-1']

# 创建互作网络图
import networkx as nx
import matplotlib.pyplot as plt

# 创建图
plt.figure(figsize=(6, 6))
G = nx.Graph()

# 添加节点和边
for _, row in network_df.iterrows():
    G.add_node(row['program'], type='program')
    G.add_node(row['guide'], type='guide')
    G.add_edge(row['program'], row['guide'], weight=row['value'])

# 计算每个guide与program的平均互作值
guide_sizes = network_df.groupby('guide')['value'].mean().to_dict()

# 绘制网络图
guide_pos = nx.circular_layout(G)  # guide点维持圆形排布
program_pos = {node: (0.5 * np.cos(2 * np.pi * i / len(network_df['program'].unique())), 
                      0.5 * np.sin(2 * np.pi * i / len(network_df['program'].unique()))) 
                for i, node in enumerate(network_df['program'].unique())}  # program点放在圆周上

# 合并位置
pos = {**guide_pos, **program_pos}

edges = G.edges(data=True)

# 根据边的权重设置线的粗细和颜色
weights = [edge[2]['weight'] for edge in edges]

weights_non_linear = np.where(np.abs(weights) >= np.percentile(np.abs(weights), 20), weights, 0)

colors = ['darkred' if weight > 0 else 'darkblue' for weight in weights]

# 绘制节点，使用guide的平均互作值来设置节点大小
nx.draw_networkx_nodes(G, pos, 
                       node_size=[guide_sizes.get(node, 1e-4) * 1e7 for node in G.nodes()],
                       node_color=['lightsteelblue' if G.nodes[node]['type'] == 'guide' else 'wheat' for node in G.nodes()])
nx.draw_networkx_edges(G, pos, width=np.abs(weights_non_linear) * 2e4, alpha=0.9, edge_color=colors)
nx.draw_networkx_labels(G, pos, font_size=7)

# 添加图例
red_patch = plt.Line2D([0], [0], marker='o', color='w', label='Positively Enriched', markerfacecolor='darkred', markersize=10)
blue_patch = plt.Line2D([0], [0], marker='o', color='w', label='Negatively Enriched', markerfacecolor='darkblue', markersize=10)
plt.legend(handles=[red_patch, blue_patch], loc='upper right', bbox_to_anchor=(1.05, 1), facecolor='white', fontsize='small', frameon=False)

sns.despine(left=True, bottom=True)
plt.title('Program-Guide Enrichment Network')

plt.rcParams['pdf.fonttype'] = 42
plt.savefig('./plots/program_guide_enrichment_network.pdf', bbox_inches='tight', format='pdf')
plt.show()
```

```python
import numpy as np
from scipy.stats import pearsonr
from concurrent.futures import ThreadPoolExecutor
from tqdm import tqdm

SAMPLE = '2-1'

# Function to compute correlations for a given row i
def compute_correlation(i, gene_matrix, guide_matrix, n_genes):
    row_corr = np.zeros(n_genes)
    for j in range(n_genes):
        row_corr[j] = pearsonr(guide_matrix[:, i], gene_matrix[:, j])[0]
    return i, row_corr

# Number of genes
n_genes = len(hvg_fdata.var_names)
n_guides = len(gdata.var_names)
gene_matrix = hvg_fdata[hvg_fdata.obs.marker == SAMPLE, :].X.toarray()
guide_matrix = gdata[gdata.obs.marker == SAMPLE, :].X
corr_matrix = np.zeros([n_guides, n_genes])
```

```python
# Use ThreadPoolExecutor with 16 threads
with ThreadPoolExecutor(max_workers=8) as executor:
    futures = [executor.submit(compute_correlation, i, gene_matrix, guide_matrix, n_genes) for i in range(n_guides)]
    
    # Collect results with progress bar
    for future in tqdm(futures, total=n_guides):
        i, row_corr = future.result()
        corr_matrix[i, :] = row_corr

print("Parallel computation completed successfully!")
```
