```python
import scanpy as sc
```

```python
fdata = sc.read_h5ad('./RNA/Multitime.filtered.tissue.h5')
```

```python
fdata.obs
```

```python
plot_data = fdata[fdata.obs.marker == 'B924-1'].copy()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

x_range = (plot_data.obsm['spatial'][:, 0].max() - plot_data.obsm['spatial'][:, 0].min()) / 2000
y_range = (plot_data.obsm['spatial'][:, 1].max() - plot_data.obsm['spatial'][:, 1].min()) / 2000

plt.figure(figsize=(x_range, y_range))
sns.scatterplot(x=plot_data.obsm['spatial'][:, 0], y=plot_data.obsm['spatial'][:, 1],
                c=plot_data[:, 'Cd44'].X.toarray().flatten(), s=10, marker='s', edgecolor='none')

sns.despine(left=True, bottom=True)
plt.xticks([])
plt.yticks([])
plt.gca().invert_yaxis()
plt.title('Cd44')
plt.show()
```
