```python
import utils

fdata = ad.read_h5ad("../20240502-SPACseq/RNA/Multitime.filtered.rna.h5")
gdata = sc.read_h5ad('../20240502-SPACseq/RNA/Multitime.filtered.guide.bin100.h5')
gdata = utils.combine_guide_replicates(gdata)
gdata.obs.marker = gdata.obs.marker.map({
    'B926-1': '3-1',
    'B926-2': '3-2',
    'B924-1': '2-1',
    'B924-2': '2-2',
    'B922-1': '1-1',
    'B922-2': '1-2'
})
fdata.obs['coord_x'] = fdata.obsm['spatial'][:, 0] // 100 * 100
fdata.obs['coord_y'] = fdata.obsm['spatial'][:, 1] // 100 * 100

gdata.obs['coord_x'] = gdata.obsm['spatial'][:, 0] // 100 * 100
gdata.obs['coord_y'] = gdata.obsm['spatial'][:, 1] // 100 * 100
fdata.obs_names = [str(marker) + '_' + str(x) + '_' + str(y) for marker, x, y in zip(fdata.obs['marker'], fdata.obs['coord_x'], fdata.obs['coord_y'])]
gdata.obs_names = [str(marker) + '_' + str(x) + '_' + str(y) for marker, x, y in zip(gdata.obs['marker'], gdata.obs['coord_x'], gdata.obs['coord_y'])]

common_bins = np.intersect1d(fdata.obs_names, gdata.obs_names)
fdata = fdata[common_bins].copy()
gdata = gdata[common_bins].copy()
```

```python
fdata.obs["cluster"] = autok.predict(fdata, use_rep="X_cellcharter", k=11)
gdata.obs["cluster"] = fdata.obs["cluster"].tolist()
```

```python
from scipy.stats import chi2_contingency
import seaborn as sns
from adjustText import adjust_text

guide_data = gdata.copy()
c_df = pd.DataFrame(guide_data.X, columns=guide_data.var_names)
c_df["cluster"] = guide_data.obs["cluster"].tolist()
c_df = c_df.groupby(["cluster"]).sum().unstack()
m_df = c_df.unstack()

ntc_cnt = m_df.loc['sgnon-targeting'].sum()
m_df = m_df.div(m_df.sum(axis=1), axis=0) * ntc_cnt
chi_dict = {}
for guide in gdata.var_names:
    if m_df.loc[guide].sum() == 0:
        chi_dict[guide] = 1
        continue
    if guide == 'sgnon-targeting': continue
    chi_dict[guide] = chi2_contingency(m_df.loc[['sgnon-targeting', guide]].T).pvalue

pdf = pd.DataFrame(chi_dict, index=['Chi p-value']).T.sort_values(by='Chi p-value', ascending=False).reset_index()
pdf.columns = ['Guide', 'pvalue']
pdf.pvalue = -np.log10(pdf.pvalue)
pdf = pdf.sort_values(by='pvalue', ascending=False)
pdf['index'] = list(range(1, pdf.shape[0]+1))

plt.figure(figsize=(3, 3))
sns.scatterplot(pdf, x='index', y='pvalue', c='black', s=10)

top_10 = pdf.head(10)
texts = []
for i, row in top_10.iterrows():
    texts.append(plt.text(row['index'], row['pvalue'], row['Guide'], fontsize=6))
adjust_text(texts, arrowprops=dict(arrowstyle='-', color='gray', lw=0.5))

plt.xticks(rotation=90)
# plt.gca().set_yscale('log')
plt.title('Chi2 p-value')
plt.xlabel('Rank')
plt.ylabel('pvalue')

sns.despine()
```

```python
from scipy.stats import spearmanr
from tqdm import tqdm

corr_matrix = np.zeros((9, 9))
for n_clusters in tqdm(range(7, 16)):
    fdata.obs["cluster"] = autok.predict(fdata, use_rep="X_cellcharter", k=n_clusters)
    gdata.obs["cluster"] = fdata.obs["cluster"].tolist()

    guide_data = gdata.copy()
    c_df = pd.DataFrame(guide_data.X, columns=guide_data.var_names)
    c_df["cluster"] = guide_data.obs["cluster"].tolist()
    c_df = c_df.groupby(["cluster"]).sum().unstack()
    m_df = c_df.unstack()

    ntc_cnt = m_df.loc['sgnon-targeting'].sum()
    m_df = m_df.div(m_df.sum(axis=1), axis=0) * ntc_cnt
    chi_dict = {}
    for guide in gdata.var_names:
        if m_df.loc[guide].sum() == 0:
            chi_dict[guide] = 1
            continue
        if guide == 'sgnon-targeting': continue
        chi_dict[guide] = chi2_contingency(m_df.loc[['sgnon-targeting', guide]].T).pvalue

    pdf = pd.DataFrame(chi_dict, index=['Chi p-value']).T.sort_values(by='Chi p-value', ascending=False).reset_index()
    pdf.columns = ['Guide', 'pvalue']
    pdf.pvalue = -np.log10(pdf.pvalue)
    pdf = pdf.sort_values(by='pvalue', ascending=False)
    pdf['index'] = list(range(1, pdf.shape[0]+1))

    for n_clusters_ in range(n_clusters, 16):
        fdata.obs["cluster"] = autok.predict(fdata, use_rep="X_cellcharter", k=n_clusters_)
        gdata.obs["cluster"] = fdata.obs["cluster"].tolist()

        guide_data = gdata.copy()
        c_df = pd.DataFrame(guide_data.X, columns=guide_data.var_names)
        c_df["cluster"] = guide_data.obs["cluster"].tolist()
        c_df = c_df.groupby(["cluster"]).sum().unstack()
        m_df = c_df.unstack()

        ntc_cnt = m_df.loc['sgnon-targeting'].sum()
        m_df = m_df.div(m_df.sum(axis=1), axis=0) * ntc_cnt
        chi_dict = {}
        for guide in gdata.var_names:
            if m_df.loc[guide].sum() == 0:
                chi_dict[guide] = 1
                continue
            if guide == 'sgnon-targeting': continue
            chi_dict[guide] = chi2_contingency(m_df.loc[['sgnon-targeting', guide]].T).pvalue

        pdf_2 = pd.DataFrame(chi_dict, index=['Chi p-value']).T.sort_values(by='Chi p-value', ascending=False).reset_index()
        pdf_2.columns = ['Guide', 'pvalue']
        pdf_2.pvalue = -np.log10(pdf_2.pvalue)
        pdf_2 = pdf_2.sort_values(by='pvalue', ascending=False)
        pdf_2['index'] = list(range(1, pdf_2.shape[0]+1))

        corr_matrix[n_clusters-7, n_clusters_-7] = spearmanr(pdf.sort_values(by='Guide')['index'], pdf_2.sort_values(by='Guide')['index'])[0]
```

```python
np.quantile(corr_matrix, 1)
```

```python
# Create a mask for the lower triangle
mask = np.tril(np.ones_like(corr_matrix, dtype=bool))
for i in range(9):
    mask[i, i] = False

# Create the heatmap with the upper triangle displayed
plt.figure(figsize=(6, 5))
ax = sns.heatmap(corr_matrix, mask=mask, cmap="Reds", annot=True, fmt='.2f', cbar=True)
ax.tick_params(left=False, bottom=False, labelright=True, labeltop=True, labelleft=False, labelbottom=False, right=True, top=True)
ax.set_xticklabels(range(7, 16))  # Add labels 7 to 15 for the top axis
ax.set_yticklabels(range(7, 16))  # Add labels 7 to 15 for the right axis

# Customize the colorbar
cbar = ax.collections[0].colorbar
cbar.ax.set_position([0.2, 0.2, 0.2, 0.2])  # Adjust the position of the colorbar (left, bottom, width, height)
cbar.set_ticks([0.929, 1])  # Set custom ticks for the colorbar
cbar.set_ticklabels([0.93, 1.00])  # Customize the tick labels
cbar.ax.tick_params(labelsize=10, width=0.5)  # Customize tick size and width

ax.set_xlabel('cluster count', loc='right')
ax.set_title("Spearman's R")

# Show the plot
plt.show()
```
