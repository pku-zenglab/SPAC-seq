# Cellcharter

```python
import scvi
import squidpy as sq
import cellcharter as cc
import numpy as np

fdata = sc.read_h5ad("/home/wpy/stereoseq/20240502-SPACseq/RNA/B926-1.downsample_3.h5")
model = scvi.model.SCVI.load("/home/wpy/stereoseq/20240502-SPACseq/scripts/scvi.downsample_3.model")
fdata.obsm['X_scVI'] = model.get_latent_representation(fdata).astype(np.float32)
sq.gr.spatial_neighbors(fdata, coord_type='generic', delaunay=True)
cc.gr.remove_long_links(fdata)
cc.gr.aggregate_neighbors(fdata, n_layers=3, use_rep='X_scVI', out_key='X_cellcharter')
```

```python
autok = cc.tl.ClusterAutoK(
    n_clusters = (7,20),
    max_runs = 10,
    model_params = dict(
        random_state = 114514,
        trainer_params = {
            'accelerator': 'gpu',
            'devices': 1
        }
    )
)
autok.fit(fdata, use_rep="X_cellcharter")
```

```python
cc.pl.autok_stability(autok)
```

```python
givenk = cc.tl.Cluster(
    n_clusters=12
)
givenk.fit(fdata, use_rep="X_cellcharter")
```

```python
fdata.obs['cellcharter_cluster'] = givenk.predict(fdata, use_rep="X_cellcharter")
fdata.write_h5ad("/home/wpy/stereoseq/20240502-SPACseq/RNA/B926-1.downsample_3.cellcharter.h5")
```

```python
import utils

guide = sc.read_h5ad("/home/wpy/stereoseq/20240502-SPACseq/RNA/B926-1.guide.bin_100.h5")
guide_1 = guide[downsample_1.obs_names].copy()
guide_2 = guide[downsample_2.obs_names].copy()
guide_3 = guide[downsample_3.obs_names].copy()

guide_1 = utils.combine_guide_replicates(guide_1)
guide_2 = utils.combine_guide_replicates(guide_2)
guide_3 = utils.combine_guide_replicates(guide_3)
```

```python
import pandas as pd
import numpy as np

mdata = guide_1
mdata.obs['cluster'] = downsample_1.obs.cellcharter_cluster.tolist()
c_df = pd.concat([pd.DataFrame(mdata.X, columns=mdata.var_names, index=mdata.obs_names), mdata.obs.cluster], axis=1)
c_df = c_df.groupby(['cluster']).sum().unstack().unstack()

g_df = c_df.apply(lambda x: 10 ** (np.log10(x + 1) - (np.log10(x + 1).sum() / len(mdata.obs.cluster.unique()))), axis=1)
g_df.head()

from scipy.spatial.distance import euclidean
dist = g_df.apply(lambda x: euclidean(x, g_df.loc['sgnon-targeting']), axis=1).values
plot_df_1 = pd.DataFrame(dist, index=g_df.index, columns=['distance']).sort_values(by='distance', ascending=False)

plot_df_1.head()
```
