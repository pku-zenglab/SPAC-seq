```python
import pandas as pd

df = pd.read_csv('./tumor_1520.csv', header=None, index_col=0)
df.columns = ['seq', 'gene']
df['gene'] = df['gene'].str.split('_').str[0].map(lambda x: 'non-targeting' if x == 'Non' else x)
df['gene_count'] = df.groupby('gene').cumcount() + 1
df
```

```python
with open('./gene_probes_cancer.gtf', 'w') as file:
    for _, row in df.iterrows():
        file.write(f"sg{row['gene']}_{row['gene_count']}\tGUIDE\tgene\t1\t50\t.\t+\t.\tgene_id \"sg{row['gene']}_{row['gene_count']}_gene\"; gene_name \"sg{row['gene']}_{row['gene_count']}_gene\";\n")
        file.write(f"sg{row['gene']}_{row['gene_count']}\tGUIDE\ttranscript\t1\t50\t.\t+\t.\tgene_id \"sg{row['gene']}_{row['gene_count']}_gene\"; gene_name \"sg{row['gene']}_{row['gene_count']}_gene\"; transcript_id \"sg{row['gene']}_{row['gene_count']}_gene\"; transcript_name \"sg{row['gene']}_{row['gene_count']}_gene\";\n")
        file.write(f"sg{row['gene']}_{row['gene_count']}\tGUIDE\texon\t1\t50\t.\t+\t.\tgene_id \"sg{row['gene']}_{row['gene_count']}_gene\"; gene_name \"sg{row['gene']}_{row['gene_count']}_gene\"; transcript_id \"sg{row['gene']}_{row['gene_count']}_gene\"; transcript_name \"sg{row['gene']}_{row['gene_count']}_gene\"; exon_number 1;\n")
```

```python
with open('./gene_probes_cancer.fa', 'w') as file:
    for _, row in df.iterrows():
        file.write(f">sg{row['gene']}_{row['gene_count']}\nATTTTAACTTGCTATTTCTAGCTCTAAAAC{row['seq']}\n")
```

```python
import random
import string
def generate_hash_value():
    return ''.join(random.choices(string.ascii_lowercase + string.digits, k=7))

with open('./Visium_Mouse_tumor_probes.csv', 'w') as file:
    for _, row in df.iterrows():
        hash_value = generate_hash_value()
        file.write(f"sg{row['gene']}_{row['gene_count']},ATTTTAACTTGCTATTTCTAGCTCTAAAAC{row['seq']},sg{row['gene']}_{row['gene_count']}|sg{row['gene']}_{row['gene_count']}_gene|{hash_value},TRUE,unspliced\n")
```
